```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "Life Expectancy Data.csv"  # Update path if needed
df = pd.read_csv(file_path)

# Strip any whitespace from column names
df.columns = df.columns.str.strip()

# Drop rows with missing GDP or Life expectancy
subset_df = df[['GDP', 'Life expectancy']].dropna()

# Calculate Pearson correlation coefficient
correlation = subset_df['GDP'].corr(subset_df['Life expectancy'])
print(f"Correlation between GDP and Life Expectancy: {correlation:.2f}")

plt.show()
```

```python
# Plotting Life Expectancy vs GDP with regression line
plt.figure(figsize=(10, 6))
sns.regplot(data=subset_df, x='GDP', y='Life expectancy', scatter_kws={'alpha': 0.4})
plt.title(f'Life Expectancy vs GDP\n(Correlation: {correlation:.2f})', fontsize=14)
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.grid(True)
```

```python
# Select numeric features of interest
num_cols = ['GDP', 'Life expectancy', 'Schooling', 'Alcohol', 'BMI', 'Income composition of resources']
corr_df = df[num_cols].dropna()

# Correlation matrix heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Status', y='Life expectancy')
plt.title("Life Expectancy by Country Status")
plt.xlabel("Country Status")
plt.ylabel("Life Expectancy")
plt.grid(True)
plt.tight_layout()
plt.show()
```
